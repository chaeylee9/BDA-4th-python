### 시리즈 추가로 다루기

```python
manu = mpg['manufacturer']
```

```python
manu[manu=='audi']
#쿼리랑 비슷하게 사용 가능할지도?
#쿼리와의 차이점 - 조건 만족하는 해당 컬럼의 값만 불러옴
```

```python
mpg.query('manufacturer=="audi"')
#쿼리는 조건 만족하는 행 전체를 데이터프레임으로 불러옴
```

```python
mpg[mpg['manufacturer']=='audi']
#이런식으로 하면 쿼리랑 같은 결과
```
