```python
from matplotlib import font_manager, rc
import matplotlib
import platform
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

%matplotlib inline 

if platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname="C:/Windows/Fonts/malgun.ttf").get_name()
    rc('font', family = font_name)
else:
    rc('font',family='AppleGothic')
    
matplotlib.rcParams['axes.unicode_minus']=False
%matplotlib inline
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic
```

```python
import matplotlib.gridspec as gridspec
```

```python
#성별에 따른 생존
#등급에 따른 생존
#일행 여부에 따른 생존
#탑승지에 따른 생존
grid = gridspec.GridSpec(2,2)
plt.figure(figsize=(20,15))
plt.subplots_adjust(wspace=0.2, hspace=0.2)

titanic_features = ['sex', 'class', 'alone', 'embark_town']

for idx, f in enumerate(titanic_features):
    ax = plt.subplot(grid[idx])
    
    sns.countplot(x=f,
                 data=titanic,
                 hue='alive',
                 palette='pastel',
                 ax=ax)
    ax.set_title(f'Survive for {f}')
```

```python
#나이 히스토그램
#fare 히스토그램
grid = gridspec.GridSpec(1,2)
plt.figure(figsize=(20,15))
plt.subplots_adjust(wspace=0.2, hspace=0.2)

titanic_features = ['age', 'fare']

for idx, f in enumerate(titanic_features):
    ax = plt.subplot(grid[idx])
    
    if max(titanic[f]) >= 200:
        sns.distplot(titanic[f],
                     color='pink',
                     bins=10,
                     ax=ax)
        ax.set_title(f'{f}')
    else:
        sns.distplot(titanic[f],
                     color='pink',
                     bins=20,
                     ax=ax)
        ax.set_title(f'{f}')
        
```

```python
#등급별 나이의 swarmplot
#등급별 fare의 swarmplot
#등급별 parch의 swarmplot
#등급별 survived의 swarmplot
grid = gridspec.GridSpec(2,2)
plt.figure(figsize=(20,15))
plt.subplots_adjust(wspace=0.2, hspace=0.2)

titanic_features = ['age', 'fare', 'parch', 'survived']

for idx, f in enumerate(titanic_features):
    ax = plt.subplot(grid[idx])
    
    sns.swarmplot(x='class',
                  y=f,
                  data=titanic,
                  ax=ax)
    ax.set_title(f'{f}')
```

```python
#등급별 나이의 성별 barplot
#등급별 fare의 성별 barplot
grid = gridspec.GridSpec(1,2)
plt.figure(figsize=(20,15))
plt.subplots_adjust(wspace=0.2, hspace=0.2)

titanic_features = ['age', 'fare']

for idx, f in enumerate(titanic_features):
    ax = plt.subplot(grid[idx])
    
    sns.barplot(x='class',
                y=f,
                hue='sex',
                palette='pastel',
                data=titanic,
                ax=ax)
    ax.set_title(f'{f}')
```

```python
#등급별 나이의 성별 boxplot
#등급별 fare의 성별 boxplot
grid = gridspec.GridSpec(1,2)
plt.figure(figsize=(20,15))
plt.subplots_adjust(wspace=0.2, hspace=0.2)

titanic_features = ['age', 'fare']

for idx, f in enumerate(titanic_features):
    ax = plt.subplot(grid[idx])
    
    sns.boxplot(x='class',
                y=f,
                hue='sex',
                palette='pastel',
                data=titanic,
                ax=ax)
    ax.set_title(f'{f}')
```

```python
hello = titanic.query('alive == "yes"').reset_index()
```

```python
hello
```

```python
male = []
female = []

classs=['First', 'Second', 'Third']
sexx=['male', 'female']

for i in classs:
    tmp = len(hello.loc[(hello['class']==i) & (hello['sex']=='male')])
    male.append(tmp)
```

```python
male
```

```python
for i in classs:
    tmp = len(hello.loc[(hello['class']==i) & (hello['sex']=='female')])
    female.append(tmp)
```

```python
female
```

```python
hi = pd.DataFrame({'male':male,
                  'female':female},
                 index=classs)
```

```python
hi #등급과 성별에 따른 생존자 수
```

```python
alone = []
notalone = []

classs=['First', 'Second', 'Third']

for i in classs:
    tmp = len(hello.loc[(hello['class']==i) & (hello['alone']==True)])
    alone.append(tmp)
```

```python
for i in classs:
    tmp = len(hello.loc[(hello['class']==i) & (hello['alone']==False)])
    notalone.append(tmp)
```

```python
alone
```

```python
notalone
```

```python
wawa = pd.DataFrame({'alone':alone,
                  'notalone':notalone},
                 index=classs)
```

```python
wawa #등급과 일행여부에 따른 생존자 수
```

```python
#등급별 성별에 따른 생존자수 히트맵
#등급별 일행여부에 따른 생존자수 히트맵
grid = gridspec.GridSpec(1,2)
plt.figure(figsize=(20,15))
plt.subplots_adjust(wspace=0.2, hspace=0.2)

titanic_features = [hi, wawa]

for idx, f in enumerate(titanic_features):
    ax = plt.subplot(grid[idx])
    
    sns.heatmap(f, annot=True, fmt='d', cmap='Blues', ax=ax)
    ax.set_title(f'{f}')
```
