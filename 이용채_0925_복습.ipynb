### 2개 이상의 그래프 그리기

```python
mpg = pd.read_csv('mpg.csv')
```

```python
mpg
```

```python
import matplotlib.gridspec as gridspec
```

```python
grid = gridspec.GridSpec(2,2)#행개수, 열개수
plt.figure(figsize=(15, 10))
plt.subplots_adjust(wspace=0.5, hspace=0.3)

mpg_features = ['manufacturer','model','category','hwy']

for idx, feature in enumerate(mpg_features):
    #enumaerate 리스트 안에서의 인덱스와 값 모두 할당 (idx와 feature에)
    ax = plt.subplot(grid[idx]) #grid에 idx 값에 해당하는 순서에 그려라
    
    sns.countplot(x=feature, #countplot은 카테고리별로 몇개씩 있는지, feature는 mpg 열의 값 
                 data=mpg,
                 hue='drv', #카테고리별로 색상 다르게
                 palette='pastel',
                 ax=ax)#몇 번째 서브플롯에 그릴 건지
    ax.set_title(f'{feature} Distribution')
```

```python
grid = gridspec.GridSpec(2,2)#행개수, 열개수
plt.figure(figsize=(15, 10))
plt.subplots_adjust(wspace=1.0, hspace=0.7) #플롯 열 사이의 간격과 행 사이의 간격

mpg_features = ['manufacturer','model','category','hwy']

for idx, feature in enumerate(mpg_features):
    #enumaerate 리스트 안에서의 인덱스와 값 모두 할당 (idx와 feature에)
    ax = plt.subplot(grid[idx]) #grid에 idx 값에 해당하는 순서에 그려라
    
    sns.countplot(x=feature, #countplot은 카테고리별로 몇개씩 있는지, feature는 mpg 열의 값 
                 data=mpg,
                 hue='fl', #카테고리별로 색상 다르게
                 palette='pastel',
                 ax=ax)#몇 번째 서브플롯에 그릴 건지
    ax.set_title(f'{feature} Distribution')
```
