```python
qut25 = mpg['cty'].quantile(.25)
```

```python
qut75 = mpg['cty'].quantile(.75)
```

```python
iqr = qut75 - qut25
up = qut75 + iqr*1.5
under = qut25 - iqr*1.5
```

```python
mpg_new['cty'] = np.where((mpg['cty']<under)|(mpg['cty']>up), np.nan, mpg['cty'])
```

```python
name = ['displ', 'year', 'cyl', 'cty', 'hwy']
# 문자열 제외하고 결측치 처리
```

```python
for x in name:
    qut25 = mpg[x].quantile(.25)
    qut75 = mpg[x].quantile(.75)
    iqr = qut75 - qut25
    up = qut75 + iqr*1.5
    under = qut25 - iqr*1.5
    mpg[x] = np.where((mpg[x]<under)|(mpg[x]>up), np.nan, mpg[x])
```
