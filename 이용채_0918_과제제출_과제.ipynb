```python
from matplotlib import font_manager, rc
import matplotlib
import platform
import matplotlib.pyplot as plt

%matplotlib inline 


if platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname="C:/Windows/Fonts/malgun.ttf").get_name()
    rc('font', family = font_name)
else:
    rc('font',family='AppleGothic')
    
matplotlib.rcParams['axes.unicode_minus']=False
%matplotlib inline
```

```python
plt.scatter([1,2,3],[6,5,4], label ='a')
plt.scatter([-1,-2,-3],[4,5,6], label ='b')
plt.legend(loc='upper right')
plt.show()
```

```python
#색상변경
plt.scatter([1,2,3],[6,5,4], label ='a', color='cornflowerblue', s=40)
plt.scatter([-1,-2,-3],[4,5,6], label ='b', color='lightpink', s=40)
plt.legend(loc='upper right')
plt.show()
```

```python
#마커변경
plt.scatter([1,2,3],[6,5,4], label ='a', color='cornflowerblue', s=60, marker='x')
plt.scatter([-1,-2,-3],[4,5,6], label ='b', color='lightpink', s=60, marker='v')
plt.legend(loc='upper right')
plt.show()
```

```python
#선스타일 변경
plt.plot([1,2,3], [4,5,6], color='darkolivegreen', label='line a', linestyle='-.', linewidth=2)
plt.plot([1,2,3], [7,8,9], color='indigo', label='line b', linestyle=':', linewidth=3)
plt.legend(loc='center')
plt.show()
#-는 solid, --는 dashed, -.는 dash-dot, :는 dotted
```

```python
#선그래프에서 마커 변경
plt.plot([1,2,3], [4,5,6], color='darkolivegreen', label='line a', linestyle='-', 
         linewidth=2, marker='D', ms=7, mew=3, mec='pink', mfc='white')
plt.plot([1,2,3], [7,8,9], color='indigo', label='line b', linestyle='-', 
         linewidth=2, marker='s', ms=7, mew=3, mec='pink', mfc='white')
plt.legend(loc='center')
plt.show()
#약자로 사용가능
```

- color:c
- linewidth:lw
- linestyle:ls
- markersize:ms
- markeredgecolor:mec
- markeredgewidth:mew
- markerfacecolor:mfc

```python
#그리드 설정
plt.plot([1,2,3], [4,5,6], color='darkolivegreen', label='line a', linestyle='-', 
         linewidth=2, marker='D', ms=7, mew=3, mec='pink', mfc='white')
plt.plot([1,2,3], [7,8,9], color='indigo', label='line b', linestyle='-', 
         linewidth=2, marker='s', ms=7, mew=3, mec='pink', mfc='white')
plt.legend(loc='center')
plt.grid(True)
plt.show()
```

```python
#tick설정
#원하는 위치에 원하는 문자로 가능
plt.plot([1,2,3], [4,5,6], color='darkolivegreen', label='line a', linestyle='-', 
         linewidth=2, marker='D', ms=7, mew=3, mec='pink', mfc='white')
plt.plot([1,2,3], [7,8,9], color='indigo', label='line b', linestyle='-', 
         linewidth=2, marker='s', ms=7, mew=3, mec='pink', mfc='white')
plt.legend(loc='center')
plt.xticks([1.25, 1.75, 2.5], ['A', 'B', 'C'])
plt.grid(True)
plt.show()
```

```python
#여러개의 플롯 하나의 윈도우로 보여주기
f = plt.figure(figsize=(5, 10))
ax1 = f.add_subplot(2,1,1)
ax1.scatter([1,2,3],[6,5,4], label ='a', color='cornflowerblue', s=60, marker='x')
ax1.scatter([-1,-2,-3],[4,5,6], label ='b', color='lightpink', s=60, marker='v')
ax1.legend(loc='upper right')

ax2 = f.add_subplot(2,1,2)
ax2.plot([1,2,3], [4,5,6], color='darkolivegreen', label='line a', linestyle='-.', linewidth=2)
ax2.plot([1,2,3], [7,8,9], color='indigo', label='line b', linestyle=':', linewidth=3)
ax2.legend(loc='center')
```

```python
f = plt.figure(figsize=(10,5))
ax1 = f.add_subplot(121)
ax1.scatter([1,2,3],[6,5,4], label ='a', color='cornflowerblue', s=60, marker='x')
ax1.scatter([-1,-2,-3],[4,5,6], label ='b', color='lightpink', s=60, marker='v')
ax1.legend(loc='upper right')

ax2 = f.add_subplot(122)
ax2.plot([1,2,3], [4,5,6], color='darkolivegreen', label='line a', linestyle='-.', linewidth=2)
ax2.plot([1,2,3], [7,8,9], color='indigo', label='line b', linestyle=':', linewidth=3)
ax2.legend(loc='center')
```

```python
#seaborn은 주로 데이터 분포를 시각화할 때 사용
import seaborn as sns
```

```python
import pandas as pd
df = pd.DataFrame({'year':[2015, 2016, 2017, 2018, 2019],
                  'count':[50, 130, 500, 350, 200]})
```

```python
df
```

```python
sns.barplot(data=df, x='year', y='count')
```

```python
sns.distplot(df['count'])
```

```python
#예시데이터 사용해보기
f = sns.load_dataset('flights')
```

```python
sns.barplot(data=f, x='year', y='passengers')
```

```python
sns.swarmplot(data=f, x="year", y="passengers")
#stripplot과 달리 데이터를 겹치지 않게 표시해줌
```
