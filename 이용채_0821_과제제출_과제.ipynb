```python
import pandas as pd
```

```python
mpg = pd.read_csv('mpg.csv')
```

```python
mpg
```

```python
# 제조사를 기준으로 그룹화한 뒤 (index = False로), displ, cty, hwy에 대해 평균 구하기
mpg.groupby('manufacturer', as_index=False)[['displ', 'cty', 'hwy']].mean()
```

```python
# 모델명이 passat인 데이터를 trans로 그룹화한 뒤 drv count
mpg.query('model == "passat"').groupby('trans', as_index=False).agg(ct=('drv', 'count'))
```

```python
df1 = mpg[['manufacturer', 'model', 'trans']][0:10]
```

```python
df1
```

```python
df2 = mpg[['manufacturer', 'model', 'cyl']][0:10]
```

```python
df2
```

```python
display(df1)
display(df2)
```

```python
# 행을 기준으로 concat
display(pd.concat([df1, df2], axis=0))
# 열을 기준으로 concat
display(pd.concat([df1, df2], axis=1))
```

```python
# 제조사와 모델명을 기준으로 교집합 병합
pd.merge(df1, df2, how='inner', on=['manufacturer', 'model'])
# 생각한 결과와 다름, 모든 열에서 값이 같은 데이터가 많아서 그런듯 함
```

```python
# 모든 열에서 데이터가 다르게 구성해서 다시 시도
df_1 = mpg.loc[[0, 7, 100, 233], ['model', 'trans', 'category']].reset_index()
df_2 = mpg.loc[[0, 7, 200, 233], ['model', 'trans', 'cyl']].reset_index()
```

```python
df_1 = df_1.drop('index', axis=1)
df_2 = df_2.drop('index', axis=1)
```

```python
display(df_1)
display(df_2)
```

```python
# 모델명과 trans를 기준으로 교집합 병합
pd.merge(df_1, df_2, how = 'inner', on=['model', 'trans'])
```

```python
# cty와 hwy에서 결측치 생성
df3 = mpg[['manufacturer', 'cty', 'hwy']][0:20]
```

```python
df3
```

```python
df3.isna()
```

```python
import numpy as np
df3.loc[[2, 4, 6, 8, 10, 12], ['cty']] = np.nan
df3.loc[[3, 5, 7, 9, 11, 13], ['hwy']] = np.nan
```

```python
df3
```

```python
df3.isna()
```

```python
df3.isna().sum()
```

```python
# 평균으로 결측치 대체
cty_mean = df3['cty'].mean()
hwy_mean = df3['hwy'].mean()
```

```python
print(cty_mean)
print(hwy_mean)
```

```python
df3['cty'] = df3['cty'].fillna(cty_mean)
df3['hwy'] = df3['hwy'].fillna(hwy_mean)
```
