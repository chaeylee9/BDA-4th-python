```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'이름':['홍길동', '김영희', '박철수'],
                    '영어':[1, 2, 3],
                   '수학':[30, 40, 50],
                   '과학':[100, 200, 300],
                   '사회':[5, 6, 7]})

df2 = pd.DataFrame({'이름':['홍길동', '김영희', '박철수', '윤혜린'],
                    '영어':[1, 2, 3, 80],
                   'math':[30, 40, 50, 900],
                   'geo':[100, 200, 300, 4],
                   '사회':[5, 6, 7, 50]})
```

```python
display(df1)
display(df2)
```

## concat

```python
pd.concat([df1, df2], axis=1)
#열 방향으로 붙이기
#df1은 3행까지만 있으니까 NaN
```

```python
pd.concat([df1, df2], axis=0)
#이름, 영어, 사회는 열 이름이 같으니까 NaN 없이 붙여짐
#df1에는 math와 geo가 없고, df2에는 수학, 과학 열이 없으니까 NaN
```

## merge

```python
df1 = pd.DataFrame({'이름':['홍길동', '김영희', '박철수', '이미진'],
                    '영어':[1, 2, 3, 4],
                   '수학':[30, 40, 50, 60],
                   '과학':[100, 200, 300, 400],
                   '사회':[5, 6, 7, 8]})

df2 = pd.DataFrame({'이름':['홍길동', '김영희', '박철수', '윤혜린'],
                    '영어':[1, 2, 3, 4],
                   'math':[30, 40, 50, 900],
                   'geo':[100, 200, 300, 4],
                   '사회':[5, 6, 20, 8]})
```

```python
display(df1)
display(df2)
```

```python
pd.merge(df1, df2, on=['이름', '영어', '사회'])
#이름, 영어, 사회에서 중복되는 값 기준으로
```

```python
pd.merge(df1, df2, on=['영어', '사회'])
#영어는 4행 모두 같은 값이지만 사회는 박철수가 다름
```

```python
df1 = pd.DataFrame({'이름':['홍길동', '김영희', '박철수', '이미진'],
                   '수학':[30, 40, 50, 60],
                   '과학':[100, 200, 300, 400]})

df2 = pd.DataFrame({'name':['홍길동', '김영희', '박철수', '윤혜린'],
                   'math':[30, 40, 90, 900],
                   'geo':[100, 200, 300, 4]})
```

```python
display(df1)
display(df2)
```

```python
pd.merge(df1, df2, left_on=['이름', '수학'], right_on=['name', 'math'])
#이름과 수학 컬럼에서 모두 같은 홍길동과 김영희만 병합
```

```python
df1_re = df1. set_index('이름')
```

```python
df1_re
```

```python
pd.merge(df1_re, df2, left_index=True, right_on='name')
#df1_re의 인덱스와 df2의 name에서 겹치는 값 기준으로 병합
```

```python
df1 = pd.DataFrame({'이름':['홍길동', '김영희', '박철수', '이미진'],
                   '수학':[30, 40, 50, 60],
                   '과학':[100, 200, 300, 400]})

df2 = pd.DataFrame({'이름':['홍길동', '김영희', '박철수', '윤혜린'],
                   'math':[30, 40, 90, 900],
                   'geo':[100, 200, 300, 4]})
```

```python
pd.merge(df1, df2, on='이름', how='left')
```

```python
pd.merge(df1, df2, on='이름', how='right')
```

```python
pd.merge(df1, df2, on='이름', how='outer')
```
