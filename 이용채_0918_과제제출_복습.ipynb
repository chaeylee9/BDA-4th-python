### 선 그래프 그리기
plt.plot

```python
plt.plot([1,2,3], [4,5,6], label='line a')
plt.plot([1,2,3], [7,8,9], label='line b')
plt.legend(loc='center')
plt.show()
```

```python
plt.plot([1,2,3], [4,5,6], color='pink', label='line a')
plt.plot([1,2,3], [7,8,9], color='skyblue', label='line b')
plt.legend(loc='center')
plt.show()
```

```python
plt.plot([1,2,3], [4,5,6], color='pink', label='line a', linestyle='dashed', linewidth=2)
plt.plot([1,2,3], [7,8,9], color='skyblue', label='line b', linestyle='dotted', linewidth=3)
plt.legend(loc='center')
plt.show()
```

```python
plt.plot([1,2,3], [4,5,6], marker='*', markersize=7, color='pink', label='line a')
plt.plot([1,2,3], [7,8,9], marker='o', markersize=7, color='skyblue', label='line b')
plt.legend(loc='center')
plt.show()
```

#### 산점도의 마커사이즈는 s로 입력

```python
plt.scatter([1,2,3],[6,5,4], marker='*', s=20, label ='a')
plt.scatter([-1,-2,-3],[4,5,6], marker='D', s=20, label ='b')
plt.legend(loc='lower left')
plt.show()
```
