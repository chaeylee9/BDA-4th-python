```python
year_re = year.iloc[:, [0,1,2,3,4,6,14,17,20,24,26,29,31,33,35,46]]
```

```python
year_re
```

```python
melted = pd.melt(year_re, id_vars=['지점', '지점명', '일시'])
```

```python
melted
```

```python
tmp = melted['variable'].str.split('(')
```

```python
tmp
```

```python
tmp = tmp.str.get(0)
```

```python
tmp
```

```python
how = []

for i in range(0, len(tmp)):
    left = tmp[i][:2]
    how.append(left)
```

```python
how
```

```python
kind = []

for i in range(0, len(tmp)) :
    right = tmp[i][2:]
    if right[0] == ' ':
        r = right[1:]
    else:
        r = right
    kind.append(r)
```

```python
kind
```

```python
melted['집계방법'] = how
melted['관측종류'] = kind
```

```python
melted
```

```python
melted.info()
```

```python
month = []

for i in range(0, len(melted)):
    m = melted['일시'][i][:7]
    month.append(m)
```

```python
melted.groupby(['집계방법', '관측종류']).mean()
#2021, 2022년의 10, 11월 전체 평균
```

```python
melted['관측월'] = month
```

```python
melted
```

```python
melted.groupby(['관측월', '집계방법', '관측종류']).mean()
```
